### DECORATORS:

```python
#extend function by passing it into another function
#functionality of the wrapper is passed to the function when using @abc on the line above the function definition

def naming(function):
    def wrapper(*args,**kwargs):
        print(f'Using function {function.__name__}:')
        return function(*args,**kwargs)
    return wrapper

@naming
def greetings(greet_message):
    print(greet_message)
    return None

greetings('Hello!')
```

### MULTIPROCESSING:

```python
%%writefile multicpu.py
from multiprocessing import Pool, cpu_count

cores = cpu_count()
print(f"Computer has {cores} cores.")

def square_number(n):
    return n * n

if __name__ == "__main__":
    with Pool(processes=4) as pool:
        results = pool.map(square_number, [1, 2, 3, 4, 5])
    print(results)

#EXAMPLE
import time
# 1.Without Parallelism(Sequential)
def heavy_task(n):
    # Simulating a CPU-intensive calculation
    return sum(i * i for i in range(n))

x = 10**2
data = [x, x, x, x, x]

start = time.time()
results = [heavy_task(x) for x in data] # One after another
end = time.time()

print(f"Sequential time: {end - start:.2f} seconds")

# 2. With Parallelism (Multiprocessing)
if __name__ == "__main__":
    start = time.time()
    with Pool(processes=16) as pool:
        results = pool.map(heavy_task,data)
    end = time.time()
    print(f"Parallel time: {end - start:.2f} seconds")
```

```python
import sympy as sp
import IPython
import matplotlib.pyplot as plt


# sp.init_printing()

sp.sqrt(8)
sp.sqrt(32)

x,y = sp.symbols('x y')
f1 = sp.sqrt(x**2 + y**2)
f1.subs({x:3,y:4})

f2 = (x**2 + 2*x + 2)**2
f2 = sp.expand(f2)

for i in range(5):
    sp.expand((x + y)**i)

sp.solve(x**2 - 4, x)
sp.solve(x**2 - 4, x, dict=True)

sp.solve(sp.Eq(x**2, 6),x)

sp.limit(sp.sin(x)/x, x, 0)

sp.diff(sp.sin(x), x)
sp.integrate(-x**2, x)

#Differential equations:
t, k = sp.symbols('t k')
P = sp.Function('P')
sp.diff(P(t), t)
e1 = sp.Eq(sp.diff(P(t), t), k*P(t))
sp.dsolve(e1) 
#C1 is the initial population at t=0, 
# e**(k*t) is the growth rate (+-) per unit of time
# P(t) is the population at time t
# P(t) = C1*e**(k*t) is the solution to the differential equation, which describes how the population changes over time based on the growth rate k and the initial population C1.
# For doubling of the population for every unit of time, k would be ln(2) 
#  because e**(ln(2)) = 2, meaning the population doubles every unit of time.

#For non constant growth rate:
# r is the intrinsic growth rate, 
# K is the carrying capacity 
# P(t) is the population at time t
IPython.display.Math(r"\frac{dP(t)}{dt} = rP(t) \left( 1 - \frac{P(t)}{K} \right)")
t, r, K, P0 = sp.symbols('t r K P0')
P = sp.Function('P')

# Define the Logistic Equation
diff_eq = sp.Eq(P(t).diff(t), r * P(t) * (1 - P(t) / K))

# Solve it
solution = sp.dsolve(diff_eq, ics={P(0): sp.symbols('P0')})
print("General Solution: ")
solution
simplify_solution = sp.simplify(solution.rhs)
simplify_solution
specific_solution = simplify_solution.subs({r:0.05, P0:5, K:1000})
print(specific_solution.evalf())
plt.rcParams['axes.grid'] = True #add grid to all plots
sp.plot(specific_solution,(t,0,250), title="Logistic Growth", 
        ylabel="Population P(t)", xlabel="Time t",line_color='blue'
        )
```

```python
#BINARY SEARCH VS LINEAR SEARCH:
# create list
# check value in range
# check limits
# split list  
# search

import random

#Create list
def create_list(n,N):
  lst = random.sample(range(N), n)
  lst.sort()
  target = random.sample(range(N),1)[0]
  return lst, target

lst1, target = create_list(10,100)

print(f"Target: {target} \nList: {lst1}")

#Check if target is in range of the list
def in_range(lst,target):
    # print(f"""
    # First element: {lst[0]}
    # Last element: {lst[-1]}
    # Target: {target}
    # Within Range: {(lst[0]<=target<=lst[-1])}
    # """)
    return (lst[0]<=target<=lst[-1])

in_range([3],3)
in_range([1,2,3,4,5],7)
in_range([1,2,3,4,5],3)

#Check if target is in the limits of the list
def check_limits(lst,target):
  if (target == lst[0] or target == lst[-1]):
    return True
  return False

def split(lst):
    half = len(lst)//2
    return lst[:half], lst[half:]

split([1,2,3,4,5])

#Split and search:
def search(lst,target):
  s = 0
  while len(lst)>1:
    s += 1
    # print(lst)
    if check_limits(lst,target):
      return True, s
    else:
      first_half, second_half = split(lst)
      if in_range(first_half,target):
        lst = first_half
      elif in_range(second_half,target):
        lst = second_half
      else:
        return False, s
      if check_limits(lst,target):
        return True, s
  else:
    if check_limits(lst,target):
      return True, s
    else:
      return False, s


for i in [10**i for i in range(1,9)]:
  lst1, target = create_list(i,i)
  TF, s = search(lst1,target)
  print(TF, s)
```

```python
#Average number of trials to match 3,4 or 5 letters:
import string
import random
import time

from scipy.stats import norm
from joblib import Parallel, delayed
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


def generate_list(n):
    letters = list(string.ascii_lowercase)
    return random.sample(letters, n)

def trials_match_n(n):
    l1 = generate_list(n)
    trials = 0
    l2 = []
    while l1 != l2:
        l2 = generate_list(n)
        trials += 1
    return trials

n_value = 5
total_runs = 1000

print(f"Starting {total_runs} parallel runs for n={n_value}...")
start_time = time.time()

# joblib handles the backend processes automatically, even in notebooks!
# n_jobs=-1 tells it to use all available CPU cores.
results = Parallel(n_jobs=-1,return_as="generator_unordered")(delayed(trials_match_n)(n_value) for _ in range(total_runs))

end_time = time.time()

print(f"All runs completed in {end_time - start_time:.2f} seconds.")
print(f"Average trials to match: {sum(results) / total_runs:,.0f}")

# Plotting the results
plt.hist(results, bins=50, edgecolor='black')
plt.title(f"Distribution of Trials to Match {n_value} Letters")
plt.xlabel("Number of Trials")
plt.ylabel("Frequency")
plt.show()


# Histogram with Kernel Density Estimate (KDE)
sns.histplot(results, kde=True)
plt.show()

# Cumulative Count
sns.histplot(results, cumulative=True, stat="count")
plt.show()

sns.ecdfplot(results, label="Empirical (Data)", color="blue", linewidth=2)
x = np.linspace(min(results), max(results), 100)
y = norm.cdf(x, loc=np.mean(results), scale=np.std(results))
plt.plot(x, y, label="Theoretical Normal", color="red", linestyle="--")

# Formatting
plt.title("Empirical vs. Normal Cumulative Distribution")
plt.legend()
plt.show()
```

```python
#BUBBLE SORT:
# inneficient for large lists

def compare(x,y):
    return (x>y)

def bubble_sort(arr):
    passes = 0
    while True:
        passes += 1
        count = 0
        for i in range(len(arr)-1):
            if compare(arr[i], arr[i+1]):
                arr[i], arr[i+1] = arr[i+1], arr[i]
                count += 1
        if count == 0:
            print(f"Sorted in {passes} passes")
            arr = [i for i in arr]
            return arr
        else:
            continue

def arr(n, r = 100000):
    return random.sample(range(r), n)

arr1 = arr(100000)

arr1 = bubble_sort(arr1)
print(arr1)
```

```python
#MERGE SORT:
#create list out of sampling n out of r numbers
import random

def arr(n, r = 100):
    return random.sample(range(r), n)

def merge_sort(arr):
    print(f"Merge sort called with: {arr}")
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    print(f"Left: {left} \nRight: {right}\n")
    print(f"Merge: {merge(left, right)}\n")
    return merge(left, right)


def merge(left, right):
    print(f"Merging left: {left} with right: {right}")
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result


# Example usage
arr = arr(20)
print("Original:", arr)
sorted_arr = merge_sort(arr)
print("Original:", arr)
print("Sorted:  ", sorted_arr)


```

```python
import dash
from dash import dcc, html, Input, Output
import plotly.express as px
import pandas as pd

# Sample data
df = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [120, 180, 150, 220, 190, 260],
    "Profit": [30, 55, 40, 80, 60, 95]
})

app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Sales Dashboard", style={"textAlign": "center"}),

    # Dropdown to select metric
    dcc.Dropdown(
        id="metric-dropdown",
        options=[{"label": col, "value": col} for col in ["Sales", "Profit"]],
        value="Sales",
        style={"width": "50%", "margin": "auto"}
    ),

    # Chart updates based on dropdown
    dcc.Graph(id="bar-chart")
])

@app.callback(
    Output("bar-chart", "figure"),
    Input("metric-dropdown", "value")
)
def update_chart(selected_metric):
    fig = px.bar(df, x="Month", y=selected_metric,
                 title=f"Monthly {selected_metric}",
                 color=selected_metric, color_continuous_scale="Blues")
    return fig


if __name__ == "__main__":
    app.run(debug=True)
```

```python
import panel as pn
import matplotlib.pyplot as plt
import pandas as pd

pn.extension()

# Sample data
df = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [120, 180, 150, 220, 190, 260],
    "Profit": [30, 55, 40, 80, 60, 95]
})

# Widget
metric = pn.widgets.Select(name="Metric", options=["Sales", "Profit"], value="Sales")

# Reactive chart
@pn.depends(metric)
def update_chart(selected_metric):
    fig, ax = plt.subplots(figsize=(7, 4))
    ax.bar(df["Month"], df[selected_metric], color="steelblue")
    ax.set_title(f"Monthly {selected_metric}")
    ax.set_xlabel("Month")
    ax.set_ylabel(selected_metric)
    plt.tight_layout()
    plt.close(fig)          # important: prevents memory leaks
    return pn.pane.Matplotlib(fig)

# Layout
dashboard = pn.Column(
    "# Sales Dashboard",
    metric,
    update_chart
)

# dashboard.servable()
dashboard.show()
```
